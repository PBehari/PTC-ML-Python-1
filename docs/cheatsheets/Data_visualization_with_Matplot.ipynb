# The format parameter of pyplot.plot

<p>We have used "ob" in our previous example as the format parameter. It consists of two characters. The first one defines the line style or the dicrete value style, i.e. the markers, while the second one chooses a colour for the graph. The order of the two characters could have been reversed, i.e. we could have written it as "bo" as well.
If the format parameter is not given, as in the first example, the default value is "b-", i.e. a solid blue line.</p>

# Ticks in Matplotlib

```python
# xticks and yticks are methods used to get or set the locations and labels of ticks on the axes.
# - ax.get_xticks() returns the current x-axis tick positions as numeric values.
# - ax.get_xticklabels() returns the current x-axis tick labels as Text objects.
# The same applies for y-axis with get_yticks() and get_yticklabels().

# Import the matplotlib plotting module
import matplotlib.pyplot as plt

# Create a figure and axes
fig, ax = plt.subplots()

# Get the current x-axis tick positions (numeric values)
xticks = ax.get_xticks()

# Get the current x-axis tick labels (Text objects)
xticklabels = ax.get_xticklabels()

# Print the numeric x tick positions
print("xticks")
print(xticks)
print("\n")

# Print the x tick label objects
print("xticklabels")
print(xticklabels)
print("\n")

# Iterate over first 6 x tick labels and print each label object
for i in range(6):
    print(xticklabels[i])

print("\n")
print("yticks")
# Get and print the current y-axis tick positions (numeric values)
yticks = ax.get_yticks()
print(yticks)
```

# Adding Annotations in Matplotlib

```python
# linspace: It generates an array of evenly spaced numbers over a specified interval.
# This function is commonly used in data visualization and numerical computations where a range of values is required.

# Import necessary modules
import matplotlib.pyplot as plt
import numpy as np

# Generate 100 evenly spaced points between 0 and 10 (inclusive)
x = np.linspace(0, 10, 100)

# Calculate the sine of each x value
y = np.sin(x)

# Create a line plot of sin(x) vs x with a label
plt.plot(x, y, label='sin(x)')

# Find the index of the maximum value in the sine curve
max_index = np.argmax(y)

# Extract the x and y coordinates of the maximum point
max_x = x[max_index]
max_y = y[max_index]

# Add annotation pointing to the maximum point on the plot
plt.annotate(
    'Max Point',                 # Text label for the annotation
    xy=(max_x, max_y),           # Coordinates of the point to annotate
    xytext=(max_x, max_y + 0.5), # Position of the text relative to the point
    arrowprops=dict(facecolor='black', arrowstyle='->') # Style of the arrow pointing to the point
)

# Add title and axis labels
plt.title('Annotated Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')

# Display the legend for the plot
plt.legend()

# Add grid lines to the plot for easier reading
plt.grid(True)

# Show the plot window
plt.show()
```

```python
# We will demonstrate how to create two subplots side by side (in one row, two columns).
# By setting the parameter sharey='row', the subplots will share the same y-axis scale,
# and y-axis tick labels on the right subplot will be hidden to avoid repetition.

# Import the matplotlib plotting module
import matplotlib.pyplot as plt

# Optionally set the figure size for better visibility
plt.figure(figsize=(10, 5))

# Create a figure and two subplots arranged in 1 row and 2 columns
# sharey='row' means both subplots share the same y-axis scale and labels on the left only
fig, (ax1, ax2) = plt.subplots(1, 2, sharey='row')

# Add centered text "left" in the first subplot with red color and fontsize 18
ax1.text(0.5, 0.5, 
         "left",
         color="red",
         fontsize=18, 
         ha='center')

# Add centered text "right" in the second subplot with green color and fontsize 18
ax2.text(0.5, 0.5, 
         "right",
         color="green",
         fontsize=18, 
         ha='center')

# Display the figure with the two subplots
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define number of rows and columns for subplots
rows, cols = 2, 3

# Create a figure and a grid of subplots (2 rows x 3 columns)
# sharex='col': subplots in the same column share the same x-axis scale and labels
# sharey='row': subplots in the same row share the same y-axis scale and labels
fig, ax = plt.subplots(rows, cols,
                       sharex='col', 
                       sharey='row')

# Loop through each row and column index
for row in range(rows):
    for col in range(cols):
        # Add text to each subplot at position (0.5, 0.5) in axes coordinates (center)
        # Display the subplot's (row, column) coordinates as string
        ax[row, col].text(0.5, 0.5, 
                          str((row, col)),
                          color="green",
                          fontsize=18, 
                          ha='center')
```

# Visualize with Matplotlib

```python
# Pie chart

# Import modules
import numpy as np
import matplotlib.pyplot as plt

# Creating dataset
cars = ['AUDI', 'BMW', 'FORD',
		'TESLA', 'JAGUAR', 'MERCEDES']

data = [23, 17, 35, 29, 12, 41]

# Creating explode data
explode = (0.1, 0.0, 0.2, 0.3, 0.0, 0.0)

# Creating color parameters
colors = ("orange", "cyan", "brown",
		"grey", "indigo", "beige")

# Wedge properties
wp = {'linewidth': 1, 'edgecolor': "green"}

# Creating autocpt arguments
def format_value(pct, prices):
	price = int(pct / 100. * np.sum(prices))
	return "{:.2f}%\n({:d} k)".format(pct, price)

# Creating plot
fig, ax = plt.subplots(figsize=(10, 7))
wedges, texts, autotexts = ax.pie(data,
								autopct=lambda pct: format_value(pct, data),
								explode=explode,
								labels=cars,
								shadow=True,
								colors=colors,
								startangle=90,
								wedgeprops=wp,
								textprops=dict(color="magenta"))

# Adding legend
ax.legend(wedges, cars,
		title="Cars",
		loc="center left",
		bbox_to_anchor=(1.1, 0, 0.5, 1))

plt.setp(autotexts, size=10, weight="bold")
ax.set_title("Customizing pie chart")

# show plot
plt.show()
```

```python
# Column chart (Grouped bar chart) comparing coffee cups consumed last week and this week by different people

# Data: coffee cups consumed last week and this week
last_week_cups = (20, 35, 30, 35, 27)
this_week_cups = (25, 32, 34, 20, 25)

# Names corresponding to each person
names = ['Mary', 'Paul', 'Billy', 'Franka', 'Stephan']

# Create a figure with specified size and resolution (DPI)
fig = plt.figure(figsize=(6,5), dpi=200)

# Define the position and size of the axes within the figure [left, bottom, width, height]
left, bottom, width, height = 0.1, 0.3, 0.8, 0.6
ax = fig.add_axes([left, bottom, width, height]) 

# Width of each bar
width = 0.25

# Positions of the bars on the x-axis (one for each person)
ticks = np.arange(len(names))

# Plot bars for last week's coffee cups at the tick positions
ax.bar(ticks, last_week_cups, width, label='Last week')

# Plot bars for this week's coffee cups shifted by 'width' to the right to avoid overlap
ax.bar(ticks + width, this_week_cups, width, align="center", label='This week')

# Label y-axis
ax.set_ylabel('Cups of Coffee')

# Set the title of the chart
ax.set_title('Coffee Consummation')

# Set x-axis ticks to be centered between grouped bars
ax.set_xticks(ticks + width/2)

# Set x-axis tick labels to the names
ax.set_xticklabels(names)

# Add legend at the best location automatically
ax.legend(loc='best')

# Display the plot
plt.show()
```

```python
# Column chart

import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [150, 200, 250, 300]
profit = [30, 50, 70, 90]

# Number of categories
n = len(categories)

# X-axis positions
x = np.arange(n)

# Width of each bar
width = 0.35

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting sales and profit
bars1 = ax.bar(x - width/2, sales, width, label='Sales')
bars2 = ax.bar(x + width/2, profit, width, label='Profit')

# Adding labels and title
ax.set_xlabel('Products')
ax.set_ylabel('Amount')
ax.set_title('Sales and Profit by Product')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

# Displaying values on top of the bars
def add_labels(bars):
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, yval + 5, int(yval), ha='center', va='bottom')

add_labels(bars1)
add_labels(bars2)

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Column chart
import matplotlib.pyplot as plt
import numpy as np

# Sample data
np.random.seed(0)
categories = [f'Category {i}' for i in range(1, 11)]
sales = np.random.randint(100, 1000, size=len(categories))
profit = np.random.randint(-100, 300, size=len(categories))

# Normalize profit values for colormap
norm = plt.Normalize(profit.min(), profit.max())

# Create a colormap
cmap = plt.cm.RdYlGn

# Plot
fig, ax = plt.subplots(figsize=(12, 7))
bars = ax.bar(categories, sales, color=cmap(norm(profit)), edgecolor='black')

# Create a ScalarMappable and set the color map
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])

# Add a color bar
cbar = fig.colorbar(sm, ax=ax, label='Profit')

# Add labels and title
ax.set_xlabel('Categories')
ax.set_ylabel('Sales Amount')
ax.set_title('Sales Amount by Category with Profit Indication')

# Add text on bars
for bar, p in zip(bars, profit):
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, yval + 10, f'${yval}\n({p})', ha='center', va='bottom', fontsize=9)

# Show plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Bar chart
import matplotlib.pyplot as plt
import numpy as np

# Sample data
np.random.seed(0)
sales = np.random.randint(100, 1000, size=10)
profit = np.random.randint(-100, 300, size=10)
categories = [f'Product {i+1}' for i in range(len(sales))]

# Create a color map based on profit
colors = ['green' if p > 0 else 'red' for p in profit]

# Plot
plt.figure(figsize=(12, 8))
bars = plt.barh(categories, sales, color=colors, alpha=0.7, edgecolor='black')

# Add labels and title
plt.xlabel('Sales Amount')
plt.ylabel('Product')
plt.title('Sales and Profit by Product')

# Add annotations
for bar, sale, prof in zip(bars, sales, profit):
    plt.text(bar.get_width() + 10, bar.get_y() + bar.get_height()/2,
             f'Sales: {sale}\nProfit: {prof}', va='center', ha='left', color='black')

# Show plot
plt.grid(True, axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
subgroup1 = [5, 7, 8, 6]
subgroup2 = [2, 5, 6, 3]
subgroup3 = [3, 2, 4, 5]

# Creating the figure and axis
fig, ax = plt.subplots()

# Stacking the data horizontally
ax.barh(categories, subgroup1, label='Subgroup 1')
ax.barh(categories, subgroup2, left=subgroup1, label='Subgroup 2')
ax.barh(categories, subgroup3, left=np.array(subgroup1) + np.array(subgroup2), label='Subgroup 3')

# Adding labels and title
ax.set_xlabel('Values')
ax.set_ylabel('Categories')
ax.set_title('Stacked Horizontal Bar Chart Example')

# Adding the legend
ax.legend()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame
data = {
    'Sub-category': ['Furniture', 'Furniture', 'Office Supplies', 'Office Supplies', 'Technology', 'Technology'],
    'Region': ['East', 'West', 'East', 'West', 'East', 'West'],
    'Sales': [200, 150, 300, 100, 250, 180]
}

df = pd.DataFrame(data)

pivot_df = df.pivot_table(values='Sales', index='Sub-category', columns='Region', aggfunc='sum', fill_value=0)

pivot_df.plot(kind='barh', stacked=True)

# Add labels and title
plt.ylabel('Sub-category')
plt.xlabel('Sales')
plt.title('Sales by Sub-category and Region')
plt.legend(title='Region')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from datetime import datetime

# Sample data for demonstration
data = {
    'Order Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05'],
    'Sales': [100, 200, 150, 300, 250]
}

df = pd.DataFrame(data)
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Animate sales over time
df = df.groupby('Order Date').sum().reset_index()

fig, ax = plt.subplots()

bar = ax.bar(df['Order Date'][0], df['Sales'][0])
ax.set_ylim(0, df['Sales'].max() * 1.1)

def update(frame):
    ax.clear()
    ax.bar(df['Order Date'][:frame+1], df['Sales'][:frame+1])
    ax.set_ylim(0, df['Sales'].max() * 1.1)
    ax.set_title(f"Sales Over Time - Frame: {frame+1}")

ani = animation.FuncAnimation(fig, update, frames=len(df), repeat=False)

# To save the animation
ani.save('sales_animation.mp4', writer='ffmpeg')

# To display the animation
plt.show()
```
