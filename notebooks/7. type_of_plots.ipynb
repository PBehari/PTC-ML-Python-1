```python
# What are Plots?
# Plots (graphics), also known as charts, are a visual representation of data in the form of colored (mostly) graphics.
# It tells its audience the story about the data relationship through data points, lines, symbols, labels, 
# and numbers so that professionals and anyone with limited knowledge of reading data can get a fair idea of what the 
# data is trying to show. We can use the Matplotlib visualization library in Python to portray the graphs.
```

```python
# The six most commonly used Plots come under Matplotlib. These are:
#   - Line Plot
#   - Bar Plot
#   - Scatter Plot
#   - Pie Plot
#   - Histogram Plot
#   - Area Plot
```

```python
# Import packages
import matplotlib.pyplot as plt
```

```python
# Line Plot:
# Line plots connect data points with straight lines to show trends over a continuous variable (often time).
# They are ideal for visualizing how values change over intervals.
# In Matplotlib, the plot() function creates line plots.

# Create a figure and a single set of axes for plotting
fig, ax = plt.subplots()

# Plot data with years on the x-axis and corresponding amounts on the y-axis
ax.plot(['2020', '2021', '2022', '2023', '2024'], [100, 200, 50, 400, 600])

# Label the x-axis as 'Year'
plt.xlabel('Year')

# Label the y-axis as 'Amount (K)'
plt.ylabel('Amount (K)')

# Show the plot window with the line plot
plt.show()
```

```python
# Bar Plot
# Bar plots are vertical or horizontal rectangular graphs used to compare data.
# They show the magnitude of values along one axis (usually the Y-axis for vertical bars).
# The length of each bar corresponds to the value it represents.
# In Matplotlib, vertical bars are created with ax.bar(), while horizontal bars use ax.barh().

# Create a figure and a single set of axes
fig, ax = plt.subplots()

# Plot bars for "Carpenter" wages on specific positions with specified width and color
ax.bar([0.25, 2.25, 3.25, 5.25, 7.25],   # X positions for Carpenter bars
       [300, 400, 200, 600, 700],        # Heights (wages) for Carpenter
       label="Carpenter",                 # Legend label
       color='blue',                     # Bar color
       width=0.5)                       # Bar width

# Plot bars for "Plumber" wages at different positions with specified width and color
ax.bar([0.75, 1.75, 2.75, 3.75, 4.75],  # X positions for Plumber bars
       [50, 30, 20, 50, 60],            # Heights (wages) for Plumber
       label="Plumber",                  # Legend label
       color='green',                   # Bar color
       width=0.5)                       # Bar width

# Add legend to distinguish the two job types
plt.legend()

# Label x-axis as 'Days'
plt.xlabel('Days')

# Label y-axis as 'Wage'
plt.ylabel('Wage')

# Set the plot title
plt.title('Details')

# Display the plot
plt.show()
```

```python
# Area Plot
# Area plots (also called stack plots) show quantities that accumulate over a period or categories.
# They are similar to line plots but fill the area under the lines, often stacking multiple datasets.
# Useful for visualizing the part-to-whole relationship over time.
# In Matplotlib, the stackplot() function creates area plots.

# Create a figure and a single set of axes for plotting
fig, ax = plt.subplots()

# Define the x-axis values (days)
days = [1, 2, 3, 4, 5]

# Define two datasets representing different categories
age = [63, 81, 52, 22, 37]
weight = [17, 28, 72, 52, 32]

# Create dummy plots for the legend with matching colors and labels
ax.plot([], [], color='blue', label='Weather Predicted', linewidth=5)
ax.plot([], [], color='green', label='Weather Change happened', linewidth=5)

# Create the stacked area plot using the two datasets over the days
ax.stackplot(days, age, weight, colors=['blue', 'green'])

# Label the x-axis to indicate it shows time fluctuations
plt.xlabel('Fluctuation with time')

# Label the y-axis to indicate it shows days (or corresponding quantity)
plt.ylabel('Days')

# Set the plot title
plt.title('Weather report using Area Plot')

# Display the legend to identify each area
plt.legend()

# Show the plot
plt.show()
```

```python
# Scatter Plot
# Scatter plots visualize the relationship between two variables by displaying data points.
# Each point represents a pair of values: one from the independent variable (x-axis) and one from the dependent variable (y-axis).
# This helps identify correlations, clusters, or patterns in data.

# Create a figure and a single set of axes for plotting
fig, ax = plt.subplots()

# Define data for study hours (independent variable) and grades (dependent variable)
study_hours = [2, 5, 1, 0, 3, 4]
grades = [60, 95, 40, 20, 70, 75]

# Create the scatter plot with study hours on x-axis and grades on y-axis
ax.scatter(study_hours, grades)

# # Plot a line connecting the points in the order they appear
# ax.plot(study_hours, grades, color='orange', linestyle='-', linewidth=2, label='Trend Line')

# Set the plot title
plt.title('Study hours vs Grades')

# Label the x-axis as 'Study hours'
plt.xlabel('Study hours')

# Label the y-axis as 'Grades'
plt.ylabel('Grades')

# Enable grid lines for easier visualization
plt.grid(True)

# Display the scatter plot
plt.show()
```

```python
# Scatter Plot with Sorted Line
# This plot shows individual data points (scatter) along with a line connecting the points,
# ordered by the independent variable (study_hours) to reveal the trend more clearly.

# Create a figure and a single set of axes for plotting
fig, ax = plt.subplots()

# Original data: study hours (independent variable) and corresponding grades (dependent variable)
study_hours = [2, 5, 1, 0, 3, 4]
grades = [60, 95, 40, 20, 70, 75]

# Sort the data by study_hours so the connecting line flows in increasing order of study_hours
sorted_pairs = sorted(zip(study_hours, grades))  # Combine study_hours and grades into tuples and sort by study_hours
sorted_study_hours, sorted_grades = zip(*sorted_pairs)  # Separate the sorted tuples back into two lists

# Plot the original data points as a scatter plot
ax.scatter(study_hours, grades)

# Plot a line connecting the data points in sorted order to visualize the trend
ax.plot(sorted_study_hours, sorted_grades, color='orange', linestyle='-', linewidth=2, label='Trend Line')

# Add title and axis labels
plt.title('Study hours vs Grades')
plt.xlabel('Study hours')
plt.ylabel('Grades')

# Enable grid lines for better readability
plt.grid(True)

# Add a legend to identify the trend line
plt.legend()

# Show the plot
plt.show()
```

```python
# Pie Plot
# A pie plot is a circular chart divided into slices to illustrate numerical proportions.
# Each slice represents a category's contribution to the whole, often shown as a percentage.
# Pie charts are useful for showing relative sizes of parts within a dataset.
# In Matplotlib, the pie() function creates pie charts.

# Create a figure and a single set of axes for plotting
fig, ax = plt.subplots()

# Define the sizes of each slice (values)
slice = [25, 50, 36, 19]

# Define labels corresponding to each slice
labels = ['Neural Network', 'Data analytics', 'Quantum Computing', 'Machine Learning']

# Define colors for each slice
colors = ['blue', 'cyan', 'green', 'orange']

# Create the pie chart with customization:
ax.pie(
    slice,                # Slice sizes
    labels=labels,        # Labels for slices
    colors=colors,        # Colors for each slice
    startangle=90,        # Start angle to rotate the chart for better appearance
    shadow=True,          # Adds a shadow for 3D effect
    explode=(0.1, 0, 0, 0),  # "Explode" the first slice (offset from center)
    autopct='%1.1f%%'    # Display percentage value on each slice with 1 decimal
)

# Set the title of the pie chart
plt.title('Training Subjects')

# Show the plot
plt.show()
```

```python
# Histogram Plot
# Histograms are used to show the distribution of numerical data by grouping values into bins (intervals).
# Unlike bar plots that compare discrete categories, histograms display frequency counts of data points within ranges.
# In Matplotlib, the hist() function creates histograms.

# Create a figure and a single set of axes for plotting
fig, ax = plt.subplots()

# Sample population ages
population = [22, 55, 62, 45, 21, 22, 34, 42, 42, 4, 2, 8]

# Define the bin edges for grouping the ages
bins = [1, 10, 20, 30, 40, 50]

# Plot the histogram with specified bins and relative bar width (rwidth)
ax.hist(population, bins, rwidth=0.5)

# Label the x-axis to represent age groups
plt.xlabel('Age groups')

# Label the y-axis to represent the number of people in each group
plt.ylabel('Number of people')

# Set the title of the plot
plt.title('Histogram')

# Display the plot
plt.show()
```
