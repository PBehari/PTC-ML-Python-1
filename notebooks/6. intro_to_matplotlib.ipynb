## Adding data to Axes

```python
# Import the pandas library
import pandas as pd

# Define a dictionary with months and corresponding average temperatures
temperature_data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct"],
    "Temperature": [4.2, 5.1, 9.3, 13.5, 18.2, 22.4, 25.1, 24.7, 20.3, 14.8]
}

# Convert the dictionary into a pandas DataFrame
df_temp = pd.DataFrame(temperature_data)

# Display the DataFrame
df_temp
```

```python
# Create a figure and a single set of axes
fig, ax = plt.subplots()

# Plot temperature values against months on the axes
ax.plot(df_temp.Month, df_temp.Temperature)

# Display the plot
plt.show()
```

```python
# Define a second dictionary with months and a different set of temperature values
temperature_data2 = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct"],
    "Temperature": [14.2, 15.1, 19.3, 23.5, 28.2, 32.4, 35.1, 34.7, 30.3, 24.8]
}

# Convert the second temperature dictionary into a pandas DataFrame
df_temp2 = pd.DataFrame(temperature_data2)

# Display the second DataFrame
df_temp2
```

```python
# Create a figure and a set of axes
fig, ax = plt.subplots()

# Plot the first temperature dataset (df_temp) against months
ax.plot(df_temp.Month, df_temp.Temperature)

# Plot the second temperature dataset (df_temp2) against months on the same axes
ax.plot(df_temp2.Month, df_temp2.Temperature)

# Display the combined plot
plt.show()
```

## Customizing Plot

### Customizing Data Appearance

#### Adding Markers

```python
# Create a figure and a set of axes
fig, ax = plt.subplots()

# Plot temperature data with circle markers at each data point
ax.plot(df_temp.Month, df_temp.Temperature, marker='o')

# Display the plot
plt.show()
```

```python
# Create a figure and a set of axes
fig, ax = plt.subplots()

# Plot temperature data with triangle-down markers ('v') at each data point
ax.plot(df_temp.Month, df_temp.Temperature, marker='v')

# Display the plot
plt.show()
```

#### Setting Linestyle

```python
# Create a figure and a set of axes
fig, ax = plt.subplots()

# Plot temperature data with triangle-down markers ('v') and dashed line style ('--')
ax.plot(df_temp.Month, df_temp.Temperature, marker='v', linestyle='--')

# Display the plot
plt.show()
```

#### Eliminating lines with linestyle

```python
# Create a figure and a set of axes
fig, ax = plt.subplots()

# Plot temperature data with triangle-down markers ('v') and no connecting line between points
ax.plot(df_temp.Month, df_temp.Temperature, marker='v', linestyle='None')

# Display the plot
plt.show()
```

#### Choosing Color

```python
# Create a figure and a set of axes
fig, ax = plt.subplots()

# Plot temperature data with red triangle-down markers ('v') and dashed red line ('--')
ax.plot(df_temp.Month, df_temp.Temperature, marker='v', linestyle='--', color='r')

# Display the plot
plt.show()
```

### Customizing Axes Labels

#### Set x-axis label

```python
# Create a figure and a set of axes
fig, ax = plt.subplots()

# Plot temperature data with red triangle-down markers ('v') and dashed red line ('--')
ax.plot(df_temp.Month, df_temp.Temperature, marker='v', linestyle='--', color='r')

# Set the label for the x-axis
ax.set_xlabel("Time (Months)")

# Display the plot
plt.show()
```

#### Set y-axis label

```python
# Create a figure and a set of axes
fig, ax = plt.subplots()

# Plot temperature data with red triangle-down markers ('v') and dashed red line ('--')
ax.plot(df_temp.Month, df_temp.Temperature, marker='v', linestyle='--', color='r')

# Set the label for the x-axis
ax.set_xlabel("Time (Months)")

# Set the label for the y-axis
ax.set_ylabel("Avg Temperature")

# Display the plot
plt.show()
```

#### Adding a Title

```python
# Create a figure and a set of axes
fig, ax = plt.subplots()

# Plot temperature data with red triangle-down markers ('v') and dashed red line ('--')
ax.plot(df_temp.Month, df_temp.Temperature, marker='v', linestyle='--', color='r')

# Set the label for the x-axis
ax.set_xlabel("Time (Months)")

# Set the label for the y-axis
ax.set_ylabel("Avg Temperature")

# Set the title of the plot
ax.set_title("Weather")

# Display the plot
plt.show()
```

```python
# Import NumPy
import numpy as np

# Create a figure with a 3x2 grid of subplots (3 rows, 2 columns)
fig, ax = plt.subplots(3, 2)

# Plot random data in the subplot at row 0, column 0
ax[0, 0].plot(np.random.rand(10))

# Plot random data in the subplot at row 0, column 1 with red color
ax[0, 1].plot(np.random.rand(10), color='r')

# Display the figure with the plotted subplots
plt.show()
```

```python
# List of months to use as x-axis labels
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Temperature data for multiple cities
city_data = {
    "Paramaribo": [3, 4, 7, 10, 14, 17, 19, 19, 16, 12, 7, 4],
    "Nickerie": [1, 3, 7, 11, 16, 19, 21, 21, 17, 12, 6, 3],
    "Wanica": [6, 8, 12, 14, 18, 24, 28, 28, 24, 17, 11, 7]
}

# Create a single figure and axes
fig, ax = plt.subplots()

# Plot each city's temperature on the same axes
for city, temps in city_data.items():
    ax.plot(months, temps, label=city)

# Add title and axis labels
ax.set_title("Average Monthly Temperatures")
ax.set_xlabel("Month")
ax.set_ylabel("Temperature (°C)")

# Show the plot
plt.show()
```

```python
# Same example with multiple plots (one per city)

# Create a figure with 1 row and 3 columns of subplots, set the overall figure size
fig, axs = plt.subplots(1, 3, figsize=(15, 4))

## ----> CITY: Paramaribo
# Plot the temperature data for the current city with circle markers
axs[0].plot(months, city_data["Paramaribo"], marker='o')

# Set the title of the current subplot to the city name
axs[0].set_title("Paramaribo")

# Label the x-axis for all subplots
axs[0].set_xlabel("Month")

axs[0].set_ylabel("Temp (°C)")

## ----> CITY: Nickerie
# Plot the temperature data for the current city with circle markers
axs[1].plot(months, city_data["Nickerie"], marker='o')

# Set the title of the current subplot to the city name
axs[1].set_title("Nickerie")

# Label the x-axis for all subplots
axs[1].set_xlabel("Month")

## ----> CITY: Wanica
# Plot the temperature data for the current city with circle markers
axs[2].plot(months, city_data["Wanica"], marker='o')

# Set the title of the current subplot to the city name
axs[2].set_title("Wanica")

# Label the x-axis for all subplots
axs[2].set_xlabel("Month")

# Display the figure with all subplots
plt.show()
```

```python
# Same example with multiple plots (one per city) --> Used loop here

# Create a figure with 1 row and 3 columns of subplots, set the overall figure size
fig, axs = plt.subplots(1, 3, figsize=(15, 4))

# Loop over each city and its temperature data
for i, (city, temps) in enumerate(city_data.items()):
    # Plot the temperature data for the current city with circle markers
    axs[i].plot(months, temps, marker='o')
    
    # Set the title of the current subplot to the city name
    axs[i].set_title(city)
    
    # Label the x-axis for all subplots
    axs[i].set_xlabel("Month")
    
    # Label the y-axis only for the first subplot to avoid repetition
    if i == 0:
        axs[i].set_ylabel("Temp (°C)")

# Display the figure with all subplots
plt.show()
```

### Sharing the y-axis range

```python
# Same example with multiple plots (one subplot per city)

# Create a figure with 1 row and 3 columns of subplots,
# set the overall figure size, and share the y-axis across all subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 4), sharey=True)

# Loop through each city and its corresponding temperature data
for i, (city, temps) in enumerate(city_data.items()):
    # Plot the temperature data for the current city using circle markers
    axs[i].plot(months, temps, marker='o')
    
    # Set the title of the current subplot to the city's name
    axs[i].set_title(city)
    
    # Set the x-axis label for the current subplot
    axs[i].set_xlabel("Month")
    
    # Set the y-axis label only for the first subplot (shared y-axis makes this sufficient)
    if i == 0:
        axs[i].set_ylabel("Temperature (°C)")

# Show the final figure with all subplots displayed
plt.show()
```
